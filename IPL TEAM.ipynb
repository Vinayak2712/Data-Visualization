# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('ipl_2023_dataset.csv')
```

```python
data = data[data['Team'] != 'Unsold']
```

```python
plt.figure(figsize = (16,8))
sns.barplot(x = data['Team'].values  ,  y = data['Team'].keys()) 
plt.title('Amount Spent  By Each Team', fontsize = 18, fontweight = 'bold')
plt.xlabel('Team Names', fontsize = 16, fontweight = 'bold')
plt.ylabel('Price', fontsize = 16,fontweight = 'bold')
plt.xticks(rotation = 30)
plt.show()
```

```python
team_counts = data['Team'].value_counts()
explode = [0.2,0,0.2,0,0.2,0,0.2,0,0.2] 
plt.pie(team_counts, labels = team_counts.index , autopct ='%1.1f%%', explode = explode)
plt.title('[Distribution Of Teams]', fontsize = 22)
plt.show()

```

```python
type_values = data['Type'].value_counts().values
type_keys = data['Type'].value_counts().keys()
```

```python
plt.pie(type_values, labels = type_keys , autopct = "%.0f%%", wedgeprops = {"linewidth": 1, "edgecolor": "white"})
plt.figure(figsize = (8,8))
plt.show()
```

```python
expensive_cricketers = data.sort_values(by = 'Price Cr', ascending = False).head(10)
```

```python
plt.figure(figsize = (18,8))
sns.barplot(x = 'Player Name', y = 'Price Cr', data = expensive_cricketers)
plt.xlabel('Player Name', fontsize = 16, fontweight = 'bold')
plt.ylabel('Price in Cr', fontsize = 16, fontweight = 'bold')
plt.title('Top 10 Expensive Players in IPL', fontsize = 22, fontweight = 'bold')
plt.show()
```
