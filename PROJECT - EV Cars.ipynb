#  data visualisation

```python
plt.figure(figsize = (22,10))
sns.barplot(x = "Brand", y = "PriceEuro", data = data )
plt.xlabel(" Band Name", fontsize = 22, fontweight = 'bold')
plt.ylabel("Price In EURO", fontsize = 22, fontweight = 'bold')
plt.title("Price Of Every Brand", fontsize = 28, fontweight = 'bold')
plt.xticks(rotation = 35, fontweight = 'bold')
plt.show()
```

```python
plt.figure(figsize = (19,8))
sns.barplot(x = "Brand", y = "Range_Km", data = data)
plt.xlabel("BRAND NAME", fontsize = 16)
plt.ylabel("RANGE", fontsize = 16)
plt.title("RANGE of CARS w.r.t BRANDS", fontsize = 18, fontweight = 'bold')
plt.xticks(rotation = 35)
plt.show()
```

```python
plt.figure(figsize = (32,10))
sns.barplot(x = "Brand", y = "TopSpeed_KmH", data = data)
plt.xlabel("BRAND NAME", fontsize = 22)
plt.ylabel("TOP SPEED", fontsize = 22)
plt.title("TOP SPEED OF CARS", fontsize = 26)
plt.show()
```

```python
plt.figure(figsize = (22,8))
sns.barplot(x = "Brand", y = "Efficiency_WhKm", data = data)
plt.xlabel("BRAND NAME", fontsize = 22)
plt.ylabel("Efficiency_WhKm", fontsize = 22)
plt.title("EFFICIENCY ", fontsize = 26, fontweight = 'bold')
plt.xticks(rotation = 30)
plt.show()
```

 # Heat-map to check the correlation of the dataset

```python
data1 = data.select_dtypes(include = ['int','float'])
```

```python
plt.figure(figsize = (18,8))
sns.heatmap(data1.corr(), cmap = 'YlGnBu', annot = True, annot_kws={"size": 14})
plt.show()
```

In a heatmap, the `cmap` (colormap) parameter is used to specify the color map or color palette that determines the colors used to represent the data values in the heatmap. There are various colormaps available in Matplotlib and other visualization libraries. Here are some common colormaps that you can use for heatmaps:

1. **'viridis'**: This is a perceptually uniform colormap, meaning it is designed to be easily interpreted by the human eye and works well for representing numerical data.

2. **'plasma'**: Another perceptually uniform colormap that is often used for heatmaps. It is also designed to be visually pleasing and informative.

3. **'cividis'**: A colormap designed to be legible by those with color vision deficiencies. It is a good choice for ensuring accessibility.

4. **'inferno'**: A colormap with a dark, fiery appearance. It can be useful for emphasizing high values in your heatmap.

5. **'magma'**: Similar to 'inferno', 'magma' is a colormap with a dark, rich appearance, suitable for emphasizing high values.

6. **'coolwarm'**: A diverging colormap that transitions from cool colors (e.g., blue) for low values to warm colors (e.g., red) for high values. It's useful for highlighting both low and high values.

7. **'YlGnBu'**: A sequential colormap that transitions from yellow to green to blue. It's often used for representing increasing values.

8. **'RdBu_r'**: A diverging colormap that transitions from red to white to blue. It's useful for showing both positive and negative values around a central point.

9. **'rainbow'**: A rainbow colormap that spans a wide range of colors. It can be visually appealing but should be used with caution as it can make some details less distinguishable.

10. **'cool'**: A simple blue-to-purple colormap that can be useful for highlighting low values.
